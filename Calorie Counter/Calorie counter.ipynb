```python
df = pd.read_csv('nutrients.csv',index_col=False)
df = df.set_index('Food')
df.index = df.index.str.lower()
df.columns= df.columns.str.lower()
df.head()
```

```python
from dataclasses import dataclass
import numpy as np
import matplotlib.pyplot as plt
```

```python

#use bodyweight and calculate the intake calories and then ask if buling or cutting and calculate the protiens accordingly

#Defaults
calorie_limit =3000 #kcal
protein_limit = 180 #gms
fat_limit=80
carb_limit=80

arg1 = input("do you wish to change the goals?. Enter Y/N!")

if arg1 == "Y":
    calorie_limit = int(input("calorie limit?\n"))
    protein_limit = int(input("protien limit?\n"))
    fat_limit = int(input("fat limit?\n"))
    carb_limit = int(input("carbs limit?\n"))
else:
    print("default values set\n")

today=[]
@dataclass
class Food:
    name:str
    calories:int
    protein:int
    fat:int
    carbs:int

done=False

while not done:
    print("""
         (1) Add a new food
         (2) Visualise progress
         (q) Quit
         """)
    choice=input("Choose an option:")

    if choice=="1":
        print("Adding new food")
        arg2 = input("do you wish to check the directory?. Enter Y/N!\n")
        if arg2 == "Y":
            print("The following list of food categories are available\n")
            print(df["category"].unique(),"\n")
            arg3 = input("want to browse further?. Y/N!\n")
            if arg3 == "Y":
                cat = input("Enter the category!")
                print(df.loc[df['category'] == cat])
            else:
                print("Loading.....")
        else:
            print("Loading......")
        name = input("Name :")
        quantity = int(input("Quantity?"))
        try:
            calories = int(df.loc[name, 'calories'])*int(quantity)
        except:
            calories = 0
        try:
            proteins = int(df.loc[name, 'protein'])*int(quantity)
        except:
            calories = 0
        try:
            fats = int(df.loc[name, 'fat'])*int(quantity)
        except:
            fats = 0
        try:
            carbs = int(df.loc[name, 'carbs'])*int(quantity)
        except:
            carbs = 0
        try:
            fiber = int(df.loc[name, 'fiber'])*int(quantity)
        except:
            fiber = 0
        try:
            saturatedfat = int(df.loc[name, 'sat.fat'])*int(quantity)
        except:
            saturatedfat = 0   
        
        food= Food(name,calories, proteins,fats,carbs)
        today.append(food)
        print("Successfully added!")
        calorie_sum = int(sum(food.calories for food in today))
        protein_sum = int(sum(food.protein for food in today))
        fats_sum = int(sum(food.fat for food in today))
        carbs_sum = int(sum(food.carbs for food in today))


        fig, axs =plt.subplots(2,2) # we have 4 subplots here, axs[0,0],axs[0,1],axs[1,0],axs[1,1]
        axs[0,0].pie([protein_sum, fats_sum, carbs_sum] ,labels=["Proteins","Fats","Carbs"], autopct ="%.2f %%")   # %.2f would have been enough to print 33.33 but to print 33.33% we must use %.2f %%
        axs[0,0].set_title("Macros Distribution")
        axs[0,1].bar([0,1,2],[protein_sum, fats_sum, carbs_sum],width= 0.4,label="Current")
        axs[0,1].bar([0.4,1.4,2.4],[protein_limit, fat_limit, carb_limit],width= 0.4,label="Goals")   #0.4 is starting point to start the bar, since width is 0.4 the prev bar starts at 0 and ends at 0.4, so our next bar must start at 0.4
        axs[0,1].set_title("Macros Progress")
       # axs[0,1].xticks( [PROTEIN, FAT, CARBS], color='orange')
        axs[1,0].pie([calorie_sum,calorie_limit-calorie_sum],labels=["Calories","Remaining"],autopct ="%.2f %%")
        axs[1,0].set_title("Calories Goal Progress")
        axs[1,1].pie([protein_sum,protein_limit-protein_sum],labels=["Calories","Remaining"],autopct ="%.2f %%")
        axs[1,1].set_title("Protein Goal Progress")
       #axs[1,1].plot(list(range(len(today))) ,np.cumsum([food.calories for food in today]),label="Calories Eaten")
        #axs[1,1].plot(list(range(len(today))), [CALORIE_GOAL_LIMIT]*len(today),label="Calorie Goal")
        #axs[1,1].legend()
        #axs[1,1].set_title("Calories Goal over time")
        fig.tight_layout()
        plt.show()
    elif choice =="q":
     done=True

    else:
     print("Invalid Choice")
```
